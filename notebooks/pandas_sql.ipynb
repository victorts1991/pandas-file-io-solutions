```python
import pandas as pd
import sqlalchemy
from sqlalchemy import create_engine, MetaData, Table, inspect, text
```

```python
engine = create_engine('sqlite:///:memory:')

# Limpa o banco de dados para o teste
with engine.connect() as connection:
    connection.execute(text('DROP TABLE IF EXISTS clientes'))
    connection.execute(text('DROP TABLE IF EXISTS empregados'))
    connection.commit()
```

```python
df_clientes_banco = pd.read_csv('../data/clientes_banco.csv')

len(df_clientes_banco)
```

```python
df_clientes_banco.head()
```

```python
df_clientes_banco.to_sql('clientes', engine, index=False)

inspector = inspect(engine)
```

```python
print(inspector.get_table_names())
```

```python
query = 'SELECT * FROM clientes WHERE Categoria_de_renda = "Empregado"'
empregados = pd.read_sql(query, engine)

empregados.head()
```

```python
empregados.to_sql('empregados', con=engine, index=False)

inspector = inspect(engine)
print(inspector.get_table_names())
```
